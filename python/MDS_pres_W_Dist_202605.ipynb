```python
from sklearn.manifold import MDS
import matplotlib.pyplot as plt
import numpy as np
```

```python
W_dist = np.loadtxt("W_Compete.csv", delimiter=",")

single_idx = np.array([0, 1, 2, 3, 4])
highlight_idx = 12
combo_idx = np.array([
    i for i in range(W_dist.shape[0])
    if i not in single_idx and i != highlight_idx
])

mds = MDS(
    n_components=2,
    dissimilarity="precomputed",
    random_state=0,
    normalized_stress="auto"
)

Y_mds = mds.fit_transform(W_dist)

plt.figure(figsize=(4, 3))

# combinatorial conditions
plt.scatter(
    Y_mds[combo_idx, 0],
    Y_mds[combo_idx, 1],
    s=60,
    label="ligand mixtures"
)

# single-ligand conditions
'''
plt.scatter(
    Y_mds[single_idx, 0],
    Y_mds[single_idx, 1],
    s=120,
    marker="*",
    label="Single ligand"
)
'''

# highlighted condition: index 12
plt.scatter(
    Y_mds[highlight_idx, 0],
    Y_mds[highlight_idx, 1],
    s=120,
    marker="D",
    label="CpG-pIC"
)

# label all points by index
'''
for i in range(W_dist.shape[0]):
    plt.text(Y_mds[i, 0], Y_mds[i, 1], str(i), fontsize=8)'''


plt.xlabel("MDS 1", fontsize=14)
plt.ylabel("MDS 2", fontsize=14)
plt.title("w/ CpG-pIC competition", fontsize=14)
plt.xticks([-1,-0.5,0,0.5,1],fontsize=14)
plt.yticks([-0.8, -0.4,0,0.4], fontsize=14)

plt.axis("equal")
plt.xlim(-1.1, 1.1)
plt.ylim(-1, 0.7)
# plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import MDS

W_dist = np.loadtxt("W_Compete.csv", delimiter=",")

single_idx = np.array([0, 1, 2, 3, 4])
highlight_idx = 12

combo_idx = np.array([
    i for i in range(W_dist.shape[0])
    if i not in single_idx and i != highlight_idx
])

mds = MDS(
    n_components=2,
    dissimilarity="precomputed",
    random_state=0,
    normalized_stress="auto"
)

Y_mds = mds.fit_transform(W_dist)

fig, ax = plt.subplots(figsize=(4, 3))

# combinatorial conditions
ax.scatter(
    Y_mds[combo_idx, 0],
    Y_mds[combo_idx, 1],
    s=60,
    label="ligand mixtures"
)

# highlighted condition: index 12
ax.scatter(
    Y_mds[highlight_idx, 0],
    Y_mds[highlight_idx, 1],
    s=120,
    marker="D",
    label="CpG-pIC"
)

# No point labels are added in this unlabeled version

ax.set_xlabel("MDS 1", fontsize=14)
ax.set_ylabel("MDS 2", fontsize=14)
ax.set_title("w/ CpG-pIC competition", fontsize=14)

ax.set_xticks([-1, -0.5, 0, 0.5, 1])
ax.set_yticks([-0.8, -0.4, 0, 0.4])
ax.tick_params(axis="both", labelsize=14)

ax.axis("equal")
ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1, 0.7)
ax.set_xlabel("")
ax.set_ylabel("")
ax.set_title("")
ax.set_xticklabels([])
ax.set_yticklabels([])

# Optional: hide legend for unlabeled/source-data style plot
# ax.legend(fontsize=10)

fig.tight_layout()

# Save unlabeled version as PDF
fig.savefig(
    "MDS_CpG_pIC_competition_unlabeled.pdf",
    format="pdf",
    bbox_inches="tight"
)

plt.show()
```
